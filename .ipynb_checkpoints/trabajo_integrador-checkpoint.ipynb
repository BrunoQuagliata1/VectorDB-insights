# An치lisis de datos

A continuaci칩n haremos varios an치lisis de datos para comparar el rendimiento de nuestra base de datos vectorial (Vectra) y la base de datos relacional (PostgreSQL).

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import numpy as np

# cargamos la data de las CSV files
vectra_data = pd.read_csv('./project-vector_db/results.csv')
postgres_data = pd.read_csv('./project-sql_db/results.csv')

# Display de toda la data
display(HTML(vectra_data.to_html()))
display(HTML(postgres_data.to_html()))
```

## Comparamos los tiempos de ejecuci칩n promedio

```python
# We can start by comparing the average execution times for each database.
avg_time_vectra = vectra_data['time'].mean()
avg_time_postgres = postgres_data['time'].mean()

# Plotting the average execution times
plt.figure(figsize=(10, 6))
plt.bar(['Vectra', 'PostgreSQL'], [avg_time_vectra, avg_time_postgres], color=['blue', 'green'])
plt.title('Comparaci칩n de Tiempo de Ejecuci칩n Promedio')
plt.ylabel('Tiempo (ms)')
plt.show()
```

**Diferencia de Tiempos**: La base de datos vectorial tiene un tiempo de ejecuci칩n promedio mucho m치s bajo que la base de datos relacional.

**Dise침o y Prop칩sito de Vectra**: Vectra est치 dise침ado para ser extremadamente r치pido al manejar datos en memoria, por eso es adecuado para escenarios donde se tiene un tama침o chico de datos. Ya que el 칤ndice completo de Vectra se carga en la memoria, la b칰squeda y la recuperaci칩n son casi instant치neas, lo que explica el tiempo de ejecuci칩n m치s bajo en comparaci칩n con PostgreSQL.

**Complejidad de PostgreSQL**: Aunque PostgreSQL puede manejar vectores y realizar consultas basadas en similitud (como la similitud coseno mostrada en el c칩digo), est치 dise침ado como un sistema de gesti칩n de bases de datos relacional. Esto significa que tiene m치s capas de procesamiento y complejidad para manejar, lo que puede resultar en tiempos de ejecuci칩n m치s largos, especialmente cuando se trabaja con vectores.

**Operaciones en Memoria vs Disco**: La operaci칩n en memoria de Vectra le permite tener tiempos de respuesta muy r치pidos, mientras que PostgreSQL, que realiza operaciones en disco, puede tener tiempos de respuesta m치s lentos. Sin embargo, esto tambi칠n significa que Vectra tiene limitaciones en t칠rminos de tama침o de los datos, ya que todo el 칤ndice debe cargarse en la memoria.

**Conveniencia de Vectra para Casos Espec칤ficos**: La naturaleza en memoria de Vectra lo hace adecuado para escenarios como la inclusi칩n de un peque침o corpus de datos en una solicitud o el uso de ejemplos para entrenamiento. Sin embargo, no es ideal para escenarios a largo plazo como la memoria de un chatbot.

Las bases de datos vectoriales est치n optimizadas para calcular r치pidamente la similitud entre vectores. Esto se hace utilizando m칠tricas como la distancia euclidiana o la similitud coseno que hablamos anteriormente. por el otro lado, las bases de datos relacionales no est치n dise침adas espec칤ficamente para este prop칩sito y, por lo tanto, tuvimos que agregar algoritmos adicionales para realizar dichas tares, lo que puede ser menos eficiente.

## Comparamos la proporci칩n de respuestas correctas

```python
unique_queries_vectra = vectra_data[['query', 'answer1']].drop_duplicates()
unique_queries_postgres = postgres_data[['query', 'answer1']].drop_duplicates()

unique_queries_combined = pd.concat([unique_queries_vectra, unique_queries_postgres]).drop_duplicates()

qa_dict = dict(zip(unique_queries_combined['query'], unique_queries_combined['answer1']))

vectra_data['top_response_correct'] = vectra_data['query'].map(qa_dict) == vectra_data['answer1']
postgres_data['top_response_correct'] = postgres_data['query'].map(qa_dict) == postgres_data['answer1']

# calculamos la proporcion de las resupuestas correctas
correct_top_vectra = vectra_data['top_response_correct'].mean()
correct_top_postgres = postgres_data['top_response_correct'].mean()

# Mostramos en una grafica
plt.figure(figsize=(10, 6))
plt.bar(['Vectra', 'PostgreSQL'], [correct_top_vectra, correct_top_postgres], color=['blue', 'green'])
plt.title('Proporci칩n de Respuestas Correctas en la Primera Respuesta')
plt.ylabel('Proporci칩n')
plt.show()
```

Se observa que ambas barras est치n en 1, lo que indica un 100% de aciertos. Esto nos indica que ambos algoritmos fueron bien implementados, ya que lograron encontrar las respuestas correctas cen todos los casos. La elecci칩n entre uno u otro depender치 de otros factores, pero en cuanto a precisi칩n, ambos hicieron el trabajo bien.

## Distribuci칩n de Tiempos de Ejecuci칩n

```python
plt.figure(figsize=(10, 6))
plt.hist(vectra_data['time'], bins=20, alpha=0.5, label='Vectra')
plt.hist(postgres_data['time'], bins=20, alpha=0.5, label='PostgreSQL')
plt.title('Distribuci칩n de Tiempos de Ejecuci칩n')
plt.xlabel('Tiempo (ms)')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.show()
```

Vectra muestra tiempos de ejecuci칩n concentrados principalmente alrededor de valores bajos, cerca de los 0 ms, mientras que PostgreSQL presenta una distribuci칩n m치s dispersa, con tiempos que se extienden hasta cerca de los 60 ms. Esto sugiere que Vectra, en general, tiene tiempos de respuesta m치s r치pidos en comparaci칩n con PostgreSQL para las consultas evaluadas.
