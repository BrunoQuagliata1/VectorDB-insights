# Análisis de datos

A continuación haremos varios análisis de datos para comparar el rendimiento de nuestra base de datos vectorial (Vectra) y la base de datos relacional (PostgreSQL).

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import numpy as np

# cargamos la data de las CSV files
vectra_data = pd.read_csv('./project-vector_db/results.csv')
postgres_data = pd.read_csv('./project-sql_db/results.csv')

# Display de toda la data
display(HTML(vectra_data.to_html()))
display(HTML(postgres_data.to_html()))
```

## Comparamos los tiempos de ejecución promedio

```python
# We can start by comparing the average execution times for each database.
avg_time_vectra = vectra_data['time'].mean()
avg_time_postgres = postgres_data['time'].mean()

# Plotting the average execution times
plt.figure(figsize=(10, 6))
plt.bar(['Vectra', 'PostgreSQL'], [avg_time_vectra, avg_time_postgres], color=['blue', 'green'])
plt.title('Comparación de Tiempo de Ejecución Promedio')
plt.ylabel('Tiempo (ms)')
plt.show()
```

**Diferencia de Tiempos**: La base de datos vectorial tiene un tiempo de ejecución promedio mucho más bajo que la base de datos relacional.

**Diseño y Propósito de Vectra**: Vectra está diseñado para ser extremadamente rápido al manejar datos en memoria, por eso es adecuado para escenarios donde se tiene un tamaño chico de datos. Ya que el índice completo de Vectra se carga en la memoria, la búsqueda y la recuperación son casi instantáneas, lo que explica el tiempo de ejecución más bajo en comparación con PostgreSQL.

**Complejidad de PostgreSQL**: Aunque PostgreSQL puede manejar vectores y realizar consultas basadas en similitud (como la similitud coseno mostrada en el código), está diseñado como un sistema de gestión de bases de datos relacional. Esto significa que tiene más capas de procesamiento y complejidad para manejar, lo que puede resultar en tiempos de ejecución más largos, especialmente cuando se trabaja con vectores.

**Operaciones en Memoria vs Disco**: La operación en memoria de Vectra le permite tener tiempos de respuesta muy rápidos, mientras que PostgreSQL, que realiza operaciones en disco, puede tener tiempos de respuesta más lentos. Sin embargo, esto también significa que Vectra tiene limitaciones en términos de tamaño de los datos, ya que todo el índice debe cargarse en la memoria.

**Conveniencia de Vectra para Casos Específicos**: La naturaleza en memoria de Vectra lo hace adecuado para escenarios como la inclusión de un pequeño corpus de datos en una solicitud o el uso de ejemplos para entrenamiento. Sin embargo, no es ideal para escenarios a largo plazo como la memoria de un chatbot.

Las bases de datos vectoriales están optimizadas para calcular rápidamente la similitud entre vectores. Esto se hace utilizando métricas como la distancia euclidiana o la similitud coseno que hablamos anteriormente. por el otro lado, las bases de datos relacionales no están diseñadas específicamente para este propósito y, por lo tanto, tuvimos que agregar algoritmos adicionales para realizar dichas tares, lo que puede ser menos eficiente.

## Comparamos la proporción de respuestas correctas

```python
unique_queries_vectra = vectra_data[['query', 'answer1']].drop_duplicates()
unique_queries_postgres = postgres_data[['query', 'answer1']].drop_duplicates()

unique_queries_combined = pd.concat([unique_queries_vectra, unique_queries_postgres]).drop_duplicates()

qa_dict = dict(zip(unique_queries_combined['query'], unique_queries_combined['answer1']))

vectra_data['top_response_correct'] = vectra_data['query'].map(qa_dict) == vectra_data['answer1']
postgres_data['top_response_correct'] = postgres_data['query'].map(qa_dict) == postgres_data['answer1']

# calculamos la proporcion de las resupuestas correctas
correct_top_vectra = vectra_data['top_response_correct'].mean()
correct_top_postgres = postgres_data['top_response_correct'].mean()

# Mostramos en una grafica
plt.figure(figsize=(10, 6))
plt.bar(['Vectra', 'PostgreSQL'], [correct_top_vectra, correct_top_postgres], color=['blue', 'green'])
plt.title('Proporción de Respuestas Correctas en la Primera Respuesta')
plt.ylabel('Proporción')
plt.show()
```

Se observa que ambas barras están en 1, lo que indica un 100% de aciertos. Esto nos indica que ambos algoritmos fueron bien implementados, ya que lograron encontrar las respuestas correctas cen todos los casos. La elección entre uno u otro dependerá de otros factores, pero en cuanto a precisión, ambos hicieron el trabajo bien.

## Distribución de Tiempos de Ejecución

```python
plt.figure(figsize=(10, 6))
plt.hist(vectra_data['time'], bins=20, alpha=0.5, label='Vectra')
plt.hist(postgres_data['time'], bins=20, alpha=0.5, label='PostgreSQL')
plt.title('Distribución de Tiempos de Ejecución')
plt.xlabel('Tiempo (ms)')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.show()
```

Vectra muestra tiempos de ejecución concentrados principalmente alrededor de valores bajos, cerca de los 0 ms, mientras que PostgreSQL presenta una distribución más dispersa, con tiempos que se extienden hasta cerca de los 60 ms. Esto sugiere que Vectra, en general, tiene tiempos de respuesta más rápidos en comparación con PostgreSQL para las consultas evaluadas.
